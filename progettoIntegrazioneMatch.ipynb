```python
#Lettura il file contenente tutti i tweet relativi al match (#INTMIL, #derbydellamadoddina, #InterMilan, #DerbyMilano)
import pandas as pd
from datetime import timedelta

match=pd.read_csv(r'match.csv')
cronaca= pd.read_csv(r'CronacaPartita.csv',sep=';')
pd.set_option('display.max_rows', None)
```

```python
#converto la colonna della date in datetime
match['data'] = pd.to_datetime(match['data'], format = '%Y-%m-%dT%H:%M:%S', utc = True)
```

```python
match.sort_values(by=['data'], inplace = True)
```

```python
match.head()
```

```python
#L'orario utilizzato nello streaming live dei tweet fa riferimanto all'UTC time, un'ora indietro rispetto al fusorario italiano
match['data']=match['data'] + timedelta(hours=1)
#estraggo l'ora e il minuto di ogni tweet
match['ora_tweet']=match['data'].dt.strftime('%H:%M')
match.head()
```

```python
#raggruppo i tweet per orario (ogni minuto)
tweet_x_minuto=match.groupby('ora_tweet').count()['username']
tweet_x_minuto=tweet_x_minuto.to_frame()
tweet_x_minuto.reset_index(inplace=True)
tweet_x_minuto
```

```python
tweet_x_minuto.rename(columns={"username":"tweet_al_Minuto"}, inplace = True)
#unisco il file dei tweet per minuto con gli eventi della partita pi√π importanti
tweetpartita=tweet_x_minuto.merge(cronaca,left_on="ora_tweet",right_on="Orario", how='outer')
```
